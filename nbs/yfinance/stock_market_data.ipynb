```python
import pandas as pd
import requests
from yfinance import Ticker, Tickers
pd.set_option('display.max_columns', None)
```

#### Getting the ticker data from `sec.gov`

```python
ticker_url = "https://www.sec.gov/files/company_tickers.json"
headers = {'User-Agent': "email@address.com"}

# get all companies data
companyTickers = requests.get(
        url=ticker_url,
        headers=headers
    )
data = [v for k,v in companyTickers.json().items()]
df = pd.DataFrame(data)

def standardize_cik(cik):
    """Padding `0` to make it 10 digits and then adding `CIK` string at start""" 
    return 'CIK' + str(cik).zfill(10)
    

df['cik_str'] = df['cik_str'].apply(standardize_cik)
df
```

#### Getting ohlcv data for company tickers

```python
# Getting all tickers in a list
companyTickersList = list(df.ticker)

# Fetching the market data for tickers
data = Tickers(companyTickersList[4176:4276])
df1 = data.history(interval='1m', start='2026-07-05', auto_adjust=False)
df1
```

#### Note on `Adj Close` and interval choice

`Adj Close` behaviour depends on `auto_adjust`, not the interval:

- `auto_adjust=True` (yfinance default) folds split/div adjustment **into** `Close` and emits **no**
  separate `Adj Close` column → it shows as all-NaN and `pivot_table(dropna=True)` drops it.
- `auto_adjust=False` keeps raw `Close` **and** a separate `Adj Close`.

For **intraday** (`1m`, `1h`) `Adj Close == Close` — no split/dividend applies mid-session, so the
column is redundant. Adjustment only matters across **days**, i.e. `1d`.

| Interval | history limit | `auto_adjust` | Adj Close | use |
|----------|--------------|---------------|-----------|-----|
| `1m` | last ~8 days | `True` | drop (redundant) | live feed → `market.ohlcv.1m` |
| `1h` | last ~730 days | `True` | drop (redundant) | medium-history backfill |
| `1d` | full history | `False` | **keep** (real adjustment) | long-horizon / return features |

Rule of thumb: **`Adj Close` only for `1d`.** Select columns explicitly rather than leaning on
`dropna`, so the schema stays stable when a field is sparse.
